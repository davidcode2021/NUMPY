```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
y=np.array([2,3,7,8,6])

plt.plot(y,"*:c", ms=10)
plt.show()

```

```python
x=np.array([3,5,6,7])
y=np.array([2,4,5,8])
font={"family":"serif","color":"blue","size":20}

plt.xlabel("demand")
plt.ylabel("supply")
plt.title("demand and supply of mango",fontdict=font,loc="right")
plt.plot(x,y,linestyle=":")

plt.grid(axis="y")


plt.show()

```

```python
x=np.arange(1,10)
y=np.arange(11,20)

plt.scatter(x,y ,color='#000000')
plt.show
```

```python
x=np.array(["Transportation","Feeding","Accomodation","Wardrobe"])
y=np.array(["N4000","N3000","N10,000","N5000"])
plt.style.use("ggplot")
plt.bar(x,y,color="#fffff0", width=0.2)
plt.title("simple bar chart of allawance")
plt.show()
```

```python
plt.style.use("ggplot")
b=np.random.normal(170,10,250)
plt.hist(b,color="#000000")
plt.show()
```

```python
p=np.array([25,15,10,20,35])

mylabel=["Feeding","Transportation","Accomoadtion","wardrope","Travelling"]
myexplode=[0,0,0,0,0]         
plt.pie(p,labels=mylabel,explode=myexplode,shadow=True,startangle=90)
plt.title("A pie chart showing the allawances of a civil servant")
plt.legend(title="key",loc="lower left")
plt.show()

         
         
         
         
```
